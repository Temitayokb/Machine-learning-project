```python
### Transformation of categorical data to become numeric
df = pd.get_dummies(Abalone)
df.head()
```

```python
### Data training Process
X=df.drop('Rings', axis=1)
y=df['Rings']
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size= 0.2, shuffle=True, random_state=42)
```

```python
Random=RandomForestRegressor()
Random.fit(X=X_train,y=y_train)
```

```python
ypred=Random.predict(X_test)
```

```python
mean_absolute_error(y_test, ypred)
```

```python
scaler = StandardScaler()
X_trainscaled=scaler.fit_transform(X=X_train)
X_testscaled=scaler.transform(X_test)
```

```python
Random.fit(X=X_trainscaled,y=y_train)
```

```python
ypred=Random.predict(X_testscaled)
```

```python
mean_absolute_error(y_test, ypred)
```

```python
ridge=Ridge()
ridge.fit(X=X_train,y=y_train)
```

```python
ypred=ridge.predict(X_test)
mean_absolute_error(y_test, ypred)
```

```python
linear=LinearRegression()
linear.fit(X=X_train,y=y_train)
ypred=linear.predict(X_test)
mean_absolute_error(y_test, ypred)
```

Finally, for the identication of Abalone's age, the model trained with RandomForestRegressor is chosen
because its the one with result closeset to zero and the best performing model.
